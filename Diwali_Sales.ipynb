```python
##Importing all the libraries 

import numpy as np #help to work with array
import pandas as pd #help to work with dataframe(Tables)
import seaborn as sn #work with charts
import matplotlib.pyplot as plt # visualizing Data 
#matplotlib inline
```

#Data cleaning process

```python

pf.drop(['Status','unnamed1'],axis=1,inplace=True) # Inplace is use to save the changes
```

```python
pf.isnull()
```

```python
pf.isnull().sum()
```

```python
pf.dropna(inplace=True)
```

```python
pf.shape
```

```python
#Amount data type changed
pf['Amount'] = pf['Amount'].astype('int')
```

```python
pf['Amount'].dtypes
```

```python
pf.columns
```

```python
#useing describe() to see the fallowing detail(mean,std,min,max,etc)
pf[['Age','Orders','Amount']].describe()
```

EDA Exploratary Data Analysis

```python
#This shows who has done more shopping based on gender.
g = sn.countplot(x = 'Gender',data = pf)

#To show the label of bars
for bars in g.containers:
    g.bar_label(bars)
```

```python
#This shows who has done more shopping based on gender and spend more money on shopping.
sales_g = pf.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False) 
sn.barplot(data=sales_g , x = 'Gender', y = 'Amount')
```

```python
pf.columns
```

```python
#This shows who has done more shopping based on gender and spend more money on shopping.

age_g = sn.countplot(data = pf , x = 'Age Group' , hue = 'Gender' )
sn.set(rc={'figure.figsize' : (7,5)})
for bars in age_g.containers:
    age_g.bar_label(bars)
```

```python
#From this graph we can see that most of the buyersa are of age group between 26-36 yrs female
age_s = pf.groupby(['Age Group'] ,as_index = False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sn.set(rc={'figure.figsize' : (6,5)})
sn.barplot(data = age_s , x='Age Group' , y='Amount')
```

```python
pf.columns
```

```python
state_order = pf.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
#fig, ax = plt.subplots(figsize=(15, 5)) #this also work like the below line 
sn.set(rc={'figure.figsize' : (15,5)})
sn.barplot(data = state_order,x = 'State',y='Orders',width=0.9)
```

```python
state_sales = pf.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sn.barplot(data = state_sales,x = 'State',y = 'Amount',width=0.9)
```
